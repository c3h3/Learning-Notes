# Histogram Equalization: $$ s = T(r) = (L-1) \int_0^r P(w) dw $$

# New pdf are constant: $$ P_s(s) = P_r(r) \left|\frac{\partial r}{\partial s}\right| $$

-   ## ref: http://programmingcomputervision.com/

```python
import numpy as npi
from PIL import Image
import matplotlib.cm as cm
im = Image.open("AquaTermi_lowcontrast.jpg").convert('L')
im_array = pylab.array(im)
pylab.imshow(im_array,cmap = cm.Greys_r)
```

```python
import numpy as np

def histeq(im,nbr_bins=256):
    """ Histogram equalization of a grayscale image. """
    
    # get image histogram
    imhist,bins = np.histogram(im.flatten(),nbr_bins,normed=True)
    cdf = imhist.cumsum() # cumulative distribution function
    cdf = 255 * cdf / cdf[-1] # normalize
    # use linear interpolation of cdf to find new pixel values
    im2 = np.interp(im.flatten(),bins[:-1],cdf)
    return im2.reshape(im.shape), cdf

```

```python
im2,cdf = histeq(im_array)
```

```python
pylab.subplot(2,2,1)
pylab.imshow(im_array,cmap = cm.Greys_r)
pylab.subplot(2,2,2)
pylab.hist(im_array.flatten(),128)
pylab.subplot(2,2,3)
pylab.imshow(im2,cmap = cm.Greys_r)
pylab.subplot(2,2,4)
pylab.hist(im2.flatten(),128)

pylab.show()
```
